```python
import math
def h(x,m):
  b=[]
  for j in range(len(m)):
      b.append (m[j]*x[j])
  if sum(b) < 0:
      z= 1 - 1/(1 + math.exp(sum(b)))
  else:
      z= 1/(1 + math.exp(-sum(b)))
  return z
```

```python
def gd(x,y,alpha,iter):
  m=len(x)
  n=len(x[0])

  slope=np.zeros(n)
  slopet=np.zeros(n)

  for _ in range(iter):
    for i in range(m):
      for j in range(n):
        xt=x[i]
        yt=y[i]
        slopet[j]+=(-1/m)*(yt-h(xt,slope))*x[i][j]
    
      slope=slope-(alpha*slopet)
      print(cost(x,y,slope))
  
  return slope 
```

```python
pred_y=[]
def predict(x,y,m):
  count=0
  res=0
  for i in range(len(x)):
    t=h(x[i],m)
    if t>0.5:
      res=1 
    else:
      res=0
    if res==y[i]:
      count+=1 
    pred_y.append(res)
  print(count)
```

```python
def cost(x,y,m):
  s=0
  for i in range(len(x)):
    xt=x[i]
    yt=y[i]
    t=h(xt,m)
    if t==0:
      s+=(1-yt)*math.log(1-t)
    elif t==1:
      s+=yt*math.log(t)
    else:
      s+=yt*math.log(t)+(1-yt)*math.log(1-t)
  s=(-1/len(x))*s
  return s
```
